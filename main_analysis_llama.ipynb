# 1. Professional stereotyping

```python
import json

with open('./data/jobs.json', 'r', encoding='utf-8') as file:
    jobs_data = json.load(file)
```

```python
jobs = jobs_data['jobs']

resume_text = jobs_data['resumes']

cities = ["San Francisco, CA", "New York, NY", "Cambridge, MA", "Des Moines, SD", "Mobile, AL", "Charlotte, NC", "Yongston, OH"]
colleges = ["Stanford University", "Harvard University", "CUNY Baruch College", 
                "Carleston College", "Ohio State University", "Wichita State University"]
degrees = ["Bachelor", "MBA", "MS", "PhD"]
races = ["black","white","asian"]
genders = ["male","female"]
```

### 1.1 City Location

```python
def create_profiles_city(college, degree):
    prompt_dict_city = {}
    prompt_dict_city_plus = {}

    for job in jobs_data["jobs"].keys():
        profiles_city = []

        for c in cities:
            prfl = {"location": c, "college": college, "degree": degree}
            profiles_city.append(prfl)

        is_tech = False
        if job in ["Data Scientist", "Software Engineer"]:
            is_tech = True

        prompt_dict_city[job] = prepare_prompts_prof(job, profiles_city, resume_text[job], jobs[job], is_tech)
        prompt_dict_city_plus[job] = prepare_prompts_prof_plus(job, profiles_city, resume_text[job], jobs[job], is_tech)

    return prompt_dict_city, prompt_dict_city_plus
```

```python
prfls1, prfls1_plus = create_profiles_city("Harvard University", "Bachelor")
prfls2, prfls2_plus = create_profiles_city("Ohio State University", "Bachelor")
prfls3, prfls3_plus = create_profiles_city("Wichita State University", "Bachelor")
```

```python
#Starting Llama local server 
prmpt = prfls1['Data Scientist'][0]
model = SLMModelInstance(prmpt)
model.start_server()
```
